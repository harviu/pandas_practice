## 1. Read the data into a data frame

```python
data = pd.read_csv('pitching.csv')
```

## 2. Display all pitchers in 2015 under each team sorted by their last names

```python
df2015=data.loc[lambda df:df.yearID==2015]
teams =[] #Get all teams
for iteam in df2015['teamID']:
    sign = 0
    for t in teams:
        if iteam == t:
            sign = 1
    if sign == 0:
        teams.append(iteam)
```

```python
teamDF = {}
for t in teams:
    teamDF[t] = df2015.loc[lambda df:df.teamID==t]
    teamDF[t]=teamDF[t].sort_values(by = 'nameLast')
```

The pitchers sorted by the last name of each team are stored in the dictionary teamDF, you can use the teamDF\[teamName\] to view one specific team. 

For example:

```python
teamDF['COL']
```

## 3. Display all pitchers in 2015 sorted by their ERAs.  This allows you to identify the best and worst pitchers.
Here only shows the playerID, name, teamID and ERA for simplicity.

```python
df2015_byERA=df2015.sort_values(by='ERA',ascending=True)
df2015_byERA[['playerID','nameFirst','nameLast','teamID','ERA']]
```

## 4. Display all pitchers in 2015 under each team sorted by their ERAs

```python
teamDF_byERA = {}
for t in teams:
    teamDF_byERA[t]=teamDF[t].sort_values(by = 'ERA')
teamDF_byERA['COL'][['playerID','nameFirst','nameLast','teamID','ERA']]
```

## 5. Display all pitchers in 2015 under each team sorted by their strike outs (SO)

```python
teamDF_bySO = {}
for t in teams:
    teamDF_bySO[t]=teamDF[t].sort_values(by = 'SO',ascending = False)
teamDF_bySO['COL'][['playerID','nameFirst','nameLast','teamID','SO']]
```

## 6. Calculate the average ERA for each team, list the teams by the average ERA in a descending order.

```python
team_means = {}
for t in teams:
    team_means[t]=teamDF[t]['ERA'].mean()
team_by_ERA = pd.Series(team_means)
```

```python
team_by_ERA.sort_values(ascending=False)
```

## 7. Calculate the total number of home runs allowed (HRs) for each team, and list the team by the count in an ascending order

```python
team_HRs = {}
for t in teams:
    team_HRs[t]=teamDF[t]['HR'].sum()
team_by_HRs = (pd.Series(team_HRs)).sort_values()
team_by_HRs
```

## 8. Calculate each team's average ERA from 2011 to 2015, and list the result as a table, where each column is a year, each row is a team


```python
df_by_year = {}
team_means_yearly = {}
for y in range(2011,2016):
    team_means_yearly[y]={}
    df_by_year[y]=data.loc[lambda df:df.yearID==y]
    for t in teams:
        team_means_yearly[y][t]=(df_by_year[y].loc[lambda df:df.teamID==t])['ERA'].mean()

team_means_yearly = pd.DataFrame(team_means_yearly)
team_means_yearly
```

## 9. Calculate the average ERAs for American League (AL) and National League (NL) from 2011 to 2015, and list the result as a table, where each column is a league and each row is a year

```python
league_means_yearly = {}
league_means_yearly["American League"]={}
league_means_yearly["National League"]={}
for y in range(2011,2016):
    league_means_yearly["American League"][y]=(df_by_year[y].loc[lambda df:df.lgID=='AL'])['ERA'].mean()
    league_means_yearly["National League"][y]=(df_by_year[y].loc[lambda df:df.lgID=='NL'])['ERA'].mean()

league_means_yearly = pd.DataFrame(league_means_yearly)
league_means_yearly
```

## 10. List the pitchers who had the most wins in each of the years from 2011 to 2015

```python
pitcher_winner = {}
for y in range(2011,2016):
    pitcher_winner[y]=df_by_year[y].sort_values('W',ascending = False).iloc[0]
```

```python
pitcher_winner= pd.DataFrame(pitcher_winner)
pitcher_winner.drop('yearID').transpose()
```
